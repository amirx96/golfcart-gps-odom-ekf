### Propagation of Uncertainty

Because the equations for $v_x$ and $v_y$ depend on $\psi$, which has uncertainty, we must propagate the uncertainty. 

$$v_x = Ve*\cos(\psi) + Vn*\sin(\psi)$$

$$v_y = -Ve*\sin\psi) + Vn*\cos(\psi)$$

To propagate the uncertainty, take the partial derivatives of $v_x$ and $v_y$ with respect the the uncertain variable(s):

$\dfrac{\delta v_x}{\delta \psi} = -Ve*\sin(\psi) + Vn*\cos(\psi)$


$\dfrac{\delta v_y}{\delta \psi} = -Ve*\cos(\psi) - Vn*\sin(\psi)$


The standard deivation of $\sigma_{v_x}$ and $\sigma_{v_y}$ can then be expressed as:

$$ \sigma_{v_x}^2  = \big(\dfrac{\delta v_x}{\delta \psi}\big)^2  \sigma_{\psi}^2 $$
$$ \sigma_{v_y}^2  = \big(\dfrac{\delta v_y}{\delta \psi}\big)^2  \sigma_{\psi}^2 $$

Putting it together:

$ \sigma_{v_x}^2 = \bigg(-Ve*\sin(\psi) + Vn*\cos(\psi)\bigg)^2  \sigma_{\psi}^2$  

$ \sigma_{v_y}^2 = \bigg(-Ve*\cos(\psi) - Vn*\sin(\psi)\bigg)^2  \sigma_{\psi}^2$
