```python
import pandas
import matplotlib.pyplot as plt
import numpy as np



# Load INS data from 1.bag
ins_data_1 = pandas.read_csv('csv/1-vectornav-ins.csv')
ins_data_1 = ins_data_1.drop(columns=['.header.stamp.secs','.header.stamp.nsecs'])
ins_data_1['time'] = pandas.to_datetime(ins_data_1['time'])
ins_data_1['time'] = (ins_data_1['time'] - ins_data_1['time'][0]).astype('timedelta64[ns]').astype('int64')*1e-9
ins_data_1 = ins_data_1.set_index('time')

# Load INS data from 2.bag
ins_data_2 = pandas.read_csv('csv/2-vectornav-ins.csv')
ins_data_2 = ins_data_2.drop(columns=['.header.stamp.secs','.header.stamp.nsecs'])
ins_data_2['time'] = pandas.to_datetime(ins_data_2['time'])
ins_data_2['time'] = (ins_data_2['time'] - ins_data_2['time'][0]).astype('timedelta64[ns]').astype('int64')*1e-9
ins_data_2 = ins_data_2.set_index('time')


# Load INS data from 3.bag
ins_data_3 = pandas.read_csv('csv/3-vectornav-ins.csv')
ins_data_3 = ins_data_3.drop(columns=['.header.stamp.secs','.header.stamp.nsecs'])
ins_data_3['time'] = pandas.to_datetime(ins_data_3['time'])
ins_data_3['time'] = (ins_data_3['time'] - ins_data_3['time'][0]).astype('timedelta64[ns]').astype('int64')*1e-9
ins_data_3 = ins_data_3.set_index('time')
plt.style.use('ggplot')
```

```python
# Histogram of Pos Uncertainty
pos_uncertainty = []
pos_uncertainty += ins_data_1['.PosUncertainty'].values.tolist()
pos_uncertainty += ins_data_2['.PosUncertainty'].values.tolist()
pos_uncertainty += ins_data_3['.PosUncertainty'].values.tolist()

plt.figure()
plt.hist(pos_uncertainty,alpha=1,density=True,bins=24);
plt.xlabel('Position Uncertainty [m]')
plt.ylabel('Density');
plt.title('INS Position Uncertainty Histogram (cumulative)')

# Histogram of Yaw Uncertainty
yaw_uncertainty = []
yaw_uncertainty += ins_data_1['.YawUncertainty'].values.tolist()
yaw_uncertainty += ins_data_2['.YawUncertainty'].values.tolist()
yaw_uncertainty += ins_data_3['.YawUncertainty'].values.tolist()

plt.figure()
plt.hist(yaw_uncertainty,alpha=1,density=True,bins=24);
plt.xlabel('Heading Uncertainty [deg]')
plt.ylabel('Density');
plt.title('INS Heading Uncertainty Histogram (cumulative)');


# Histogram of Velocity Uncertainty
vel_uncertainty = []
vel_uncertainty += ins_data_1['.VelUncertainty'].values.tolist()
vel_uncertainty += ins_data_2['.VelUncertainty'].values.tolist()
vel_uncertainty += ins_data_3['.VelUncertainty'].values.tolist()

plt.figure()
plt.hist(vel_uncertainty,alpha=1,density=True,bins=24);
plt.xlabel('Velocity Uncertainty [m/s]')
plt.ylabel('Density');
plt.title('INS Velocity Uncertainty Histogram (cumulative)');
```
